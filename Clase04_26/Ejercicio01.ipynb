## Metodo Erlang

```python
import numpy as np
import random
import math

n = 1000
x = np.array([])
u = np.array([])
for i in range(n):
    x = np.append(x, random.random())

lambd = 1/x.mean()
for element in x:
    aux = -(1/lambd) * np.log(1-element)
    u = np.append(u,aux)

print(u)

```

## Metodo Beta

```python
import numpy as np
import random
import math

n = 1000
alfa = beta = 2
u = np.array([])
while (alfa + beta > 1):
    alfa = float(input("Valor de alfa"))
    beta = float(input("Valor de beta"))

for i in range(n):
    Y1 = Y2 = 2
    while Y1 + Y2 > 1:
        u1 = random.random()
        u2 = random.random()

        Y1 = u1 ** (1/alfa)
        Y2 = u2 ** (1/beta)

    x = Y1 / (Y1 + Y2)
    x = round(x,2)
    u = np.append(u, x)
print(u)
```

## Metodo Gauchy

```python
import numpy as np
import random
import math

n = 1000

alfa = 1.2
beta = 0.8

x = np.array([])
u = np.array([])
for i in range(n):
    x = np.append(x, random.random())

for element in x:
    aux = alfa - (beta/(np.tan(np.pi * element)))
    u = np.append(u,aux)

print(u)
```
