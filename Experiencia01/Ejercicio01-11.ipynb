```python
import numpy as np
import pandas as pd
import math

data = pd.read_csv("data11.csv")

def getPromedio(array):
    n = len(array)
    suma = 0
    
    for i in array:
        suma += i

    return suma/n

def getDesviacionEstandar(array):
    
    suma = 0
    promedio = array.mean()
    n = len(array)
    for i in array:
        suma += (i - promedio) ** 2
    return round(math.sqrt(suma/n),4)

def getSesgo(array):

    ses = (array.mean() - getModa(array))/ getDesviacionEstandar(array)
    return round(ses,4)

def getCurtosis(array):
    n = len(array)
    promedio = array.mean()
    sumatoria = 0
    for i in array:
        sumatoria += (i - promedio) ** 4
    G2 = (1/n) * (sumatoria / (getDesviacionEstandar(array) ** 4))
    G2 += -3
    G2 = round(G2, 3)

    return G2
    
def getCurtosisTipo(array):
    G2 = getCurtosis(array)
    if(G2 > 0):
        return f"{G2} es Leptocurtica"
    if(G2 == 0):
        return f"{G2} es Mesocurtica"
    if(G2 < 0):
        return f"{G2} es Platicurtuca"

def getModa(array):
    m = array.mode()
    return m.iat[0]

def getJarqueBera(array):
    N = len(array)
    JB = (((getSesgo(array)) ** 2) / 6) + ((getCurtosis(array)) ** 2) / 24

    return round(N * JB,4)

def getCasosEstadisticos(array, nombre):
    print(f"\n↓ Datos estadisticos de {nombre} ↓\n")

    print("Promedio =>", round(array.mean(),3))
    print("Mediana =>", array.median())
    print("Moda =>", getModa(array))
    print("Desviacion Estandar =>", getDesviacionEstandar(array))
    print("Sesgo =>", getSesgo(array))
    print("Curtosis =>", getCurtosisTipo(array))
    print("Estadistico de Jarque-Bera =>", getJarqueBera(array))
    
DP = data.iloc[:,1]
INPE = data.iloc[:,2]
PRPOLLO = data.iloc[:,3]
PRCERDO = data.iloc[:,4]
PRRES = data.iloc[:,5]

getCasosEstadisticos(DP, '"Demanda de Pollo"')
getCasosEstadisticos(INPE, '"Ingreso real por persona"')
getCasosEstadisticos(PRPOLLO, '"Precio real de pollo por libra"')
getCasosEstadisticos(PRCERDO, '"Precio real por libra de cerdo"')
getCasosEstadisticos(PRRES, '"Precio real de la carne de res por libra"')
```
