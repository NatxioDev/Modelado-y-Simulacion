Sobre un modelo en finanzas.

a) Usted desea destinar un monto de 1000 en una inversión segura que se
compone continuamente. Si el interés que reditúa esta inversión es del 8%
anual, cuanto será el capital dentro de 8 años.

```python
#Ejercicio 03 - Inciso A

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")

x = 1000
xarray = np.array([x])


for i in range(8):
    x = x * 1.08
    xarray = np.append(xarray, x)

plt.plot(xarray)
```

b) Si desea que una inversión al 8% anual tenga el valor de 10000 en 10 años,
cuanto debería invertir hoy??

```python
#Ejercicio 03 - Inciso B

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")

x10 = 10000
P = x10/((1 + 0.08) ** 10)

print(P)

x = P
xarray = np.array([x])


for i in range(10):
    x = x * 1.08
    xarray = np.append(xarray, x)
print(xarray)

```

c) Cuánto tiempo se necesita para que una inversión de A, dada una tasa r,
duplique su valor??, Y para que triplique??.

```python
#Ejercicio 03 - Inciso C - Doble

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")
alpha = 0.08
x = 1000
xarray = np.array([x])
t = np.log10(2) / np.log10(1+alpha)
print(t)
i=0
while(i<t):
    x = x * 1.08
    xarray = np.append(xarray, x)
    i+=1

plt.plot(xarray)
print(xarray)


```

```python
#Ejercicio 03 - Inciso C - Triple

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")
alpha = 0.08
x = 1000
xarray = np.array([x])
t = np.log10(3) / np.log10(1+alpha)
print(t)
i=0
while(i<t):
    x = x * 1.08
    xarray = np.append(xarray, x)
    i+=1

plt.plot(xarray)
print(xarray)
```
