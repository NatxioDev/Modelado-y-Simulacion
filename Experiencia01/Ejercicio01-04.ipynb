## Ejercicio 4
La Ley de enfriamiento de Newton establece que, en general, la temperatura de un
cuerpo en proceso de enfriamiento cambia en razón proporcional a la diferencia
entre la temperatura del cuerpo y la temperatura del ambiente. Así, si Tt es la
temperatura del cuerpo en el tiempo t y la del medio ambiente es “a”, plantee una
Ecuación en Diferencias para esta ley y resuélvala.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

TC = 35
TA = 1
k = 0.1
TCArray = np.array([TC])
for i in range(10):

    TC = (TC * (1 - k)) + (k * TA)
    TCArray = np.append(TCArray, TC)

print(TCArray)
plt.plot(TCArray)
plt.xlabel("Tiempo")
plt.ylabel("Temperatura")


    
```
