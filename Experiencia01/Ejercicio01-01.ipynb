## Ejercicio 1
Con procedimientos recursivos resolver las siguientes Ecuaciones en Diferencias
lineales de primer orden. (Considere en todos los casos que X0=1). Desarrolle un
programa para cada uno de ellos. Programe y grafique la evolución de los mismos.

```python
#Ejercicio 01 - Inciso A

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")

x = 1.0
xarray = np.array([x])


for i in range(10):
    x = x / (i+1)
    xarray = np.append(xarray, x)

plt.plot(xarray)
```

```python
#Ejercicio 01 - Inciso B

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")

x = 1.0
xarray = np.array([x])


for i in range(50):
    x = math.cos(2.48 ** i) / (x)
    xarray = np.append(xarray, x)

plt.plot(xarray)
```

```python
#Ejercicio 01 - Inciso E

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.close("all")

x = 1.0
xarray = np.array([x])


for i in range(50):
    x = -2*i*x + i*(-1**(i**2))
    xarray = np.append(xarray, x)

plt.plot(xarray)
```
