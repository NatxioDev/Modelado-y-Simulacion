# Funcion Uniforme General



```python
import numpy as np
import math as m

def getPseudoAleatorio(n):
    x = np.array([4])
    r = np.array([])
    a = 1
    c = 8
    m = 15

    for i in range(n):
        aux = ((a*x[i]) + c)
        nuevo = aux % m
        x = np.append(x,nuevo)
        r = np.append(r,(nuevo / m))

    return r

n  = int(input("Ingrese el numero de valores a generar:"))
pseudo = getPseudoAleatorio(n)

a  = int(input("Ingrese el valor de a"))
b  = int(input("Ingrese el valor de b"))

x1 = np.array([])

for element in pseudo:
    dato = a + ((b - a) * element)
    dato = round(dato,2)
    x1 = np.append(x1, dato)

print("↓ Numeros Pseudoaleatorios ↓\n",pseudo)
print("↓ Numeros generados ↓\n",x1)
```

```python
import numpy as np
import math as m

def getPseudoAleatorio(n):
    x = np.array([4])
    r = np.array([])
    a = 1
    c = 8
    m = 15

    for i in range(n):
        aux = ((a*x[i]) + c)
        nuevo = aux % m
        x = np.append(x,nuevo)
        r = np.append(r,(nuevo / m))

    return r

n  = int(input("Ingrese el numero de valores a generar:"))
pseudo = getPseudoAleatorio(n)

lambd  = int(input("Ingrese el valor de lambda"))

x1 = np.array([])

for element in pseudo:
    dato = -(np.log(1-element))/lambd
    dato = round(dato,2)
    x1 = np.append(x1, dato)

print("↓ Numeros Pseudoaleatorios ↓\n",pseudo)
print("↓ Numeros generados ↓\n",x1)
```
