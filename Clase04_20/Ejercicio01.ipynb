# Integracion Numerica de Funciones


```python
import numpy as np
import math as m

a = 0.4
b = 100
n = 10000
z = 0.5
deltaX = (b-a)/n

x1 = np.array([])

for i in range(n+1):
    aux = a + (i * deltaX)
    x1 = np.append(x1, aux)

f = np.array([])
for i in range(n+1):
    element = x1[i]
    if i == 0 or i == len(x1)-1:
        aux = (np.e**-element) * (element**(z-1)) #funcion
        f = np.append(f, aux)
    else:
        aux = 2 * (np.e**-element) * (element**(z-1)) #funcion
        f = np.append(f, aux)

area = (deltaX/2)  * f.sum()

print(round(area,4))




```
