## Metodo Congruencial Aditivo

```python
import numpy as np
x = np.array([65,89,98,3,69])
r = np.array([])
m = 10
n = 1000
for i in range(n):
    aux = (x[len(x)-1] + x[i]) % m
    x = np.append(x,aux)
    r = np.append(r,(aux/(m-1)))
print(x)
print(r)
```

## Metodo congruencial mixto

```python
import numpy as np

x = np.array([4])
r = np.array([])
a = 1
c = 8
m = 15
n = 5
for i in range(n):
    aux = ((a*x[i]) + c)
    nuevo = aux % m
    x = np.append(x,nuevo)
    r = np.append(r,(nuevo / m))

print("↓ Residuo ↓\n",x)
print("↓ Numeros aleatorios ↓\n",r)
```
