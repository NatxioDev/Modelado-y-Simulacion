```python
import math
import pandas as pd
import numpy as np
import random

n = 1

def verificarCondiciones(y_array):
    if ((y_array[0] + y_array[3] == 1) and (y_array[0] + y_array[1] == 1) and (y_array[4] + y_array[1] == 1) and (y_array[0] + y_array[2] == 1) and (y_array[2]+ y_array[4] == 1) and (y_array[3] + y_array[4] == 1)):
        return True

def calcularEcuacion(y_array):
    return ((5 * y_array[0]) + (3 * y_array[1]) + (4 * y_array[2]) + (y_array[3]) + (4 * y_array[4]))



for i in range(n):
    y_array = np.array([0.5, 0.5, 0.5, 0.5, 0.5])

    while(True):
        y_array[0] = round(random.random(),2)
        y_array[2] = round(random.random(),2)
        y_array[2] = round(random.random(),2)
        y_array[3] = round(random.random(),2)
        y_array[4] = round(random.random(),2)


        if (verificarCondiciones(y_array)):
            aux = calcularEcuacion(y_array)

            if(i == 0): 
                minimo = aux
                arrayMinimo = y_array
            if(aux < minimo):
                minimo = aux
                arrayMinimo = y_array
            break

print("Min",minimo)
print("Min",arrayMinimo)
```
