```python
import numpy as np
import math
import random

m = m = int(input("Ingrese el numero de dias a simular: "))
n = 31

array_hombres = np.array([0.45,0.35,0.15,0.05])
array_mujeres = np.array([0.25,0.55,0.15,0.05])

sum_total_hombres = np.array([])
sum_total_mujeres = np.array([])

def getRandom():
    #Beta
    alfa = 0.4
    beta = 0.4
    u = np.array([])
    Y1 = Y2 = 2
    while Y1 + Y2 > 1:
        u1 = random.random()
        u2 = random.random()
        Y1 = u1 ** (1/alfa)
        Y2 = u2 ** (1/beta)
    x = Y1 / (Y1 + Y2)
    # x = round(x,3)
    return x

def sumasDiarias(array):
    cad = ""
    for i, ele in enumerate(array, start=1):
        cad += f"Dia {i}\t"
    cad += "\n"
    for ele in array:
        cad += f"{ele}\t"
    return cad
def obtenerNumero(array, rand, n):
    acum = 0
    for i in range(n):
        if(rand >= acum and rand < array[i]+acum):
            return i
        acum += array[i]

for i in range(m):
    sum_dia_hombres = 0
    sum_dia_mujeres = 0
    for e in range(n):
        x1 = obtenerNumero(array_hombres, getRandom(), len(array_hombres))
        x2 = obtenerNumero(array_mujeres, getRandom(), len(array_mujeres))
        # x1 = obtenerNumero(array_hombres, random.random(), len(array_hombres))
        # x2 = obtenerNumero(array_mujeres, random.random(), len(array_mujeres))

        sum_dia_hombres += x1
        sum_dia_mujeres += x2

    sum_total_hombres = np.append(sum_total_hombres, sum_dia_hombres)
    sum_total_mujeres = np.append(sum_total_mujeres, sum_dia_mujeres)


print("↓ Sumas diarias ↓")
print(f"Hombres =>\n{sumasDiarias(sum_total_hombres)}")
print(f"Mujeres =>\n{sumasDiarias(sum_total_mujeres)}")
print("↓ Promedios ↓")
print("Hombres => ",sum_total_hombres.mean())
print("Mujeres => ",sum_total_mujeres.mean())
```
