# Ejercicio 01

```python
import random
import numpy as np

def generar(n):
    data = np.array([])
    for i in range(n):
        u = random.random()
        contador = 0
        sum = 0
        while sum < 1:
            contador += 1
            sum += u
        data = np.append(data,contador)
    return data.min()


def generarArray(n):
    data = np.array([])
    for i in range(n):
        data = np.append(data, generar(10))
    return data


def media(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum/len(arr)

prom = np.array([])

a = generarArray(100)
print("100 Corridas: ",a.mean())
prom = np.append(prom, a.mean())
a = generarArray(1000)
print("1000 Corridas: ",a.mean())
prom = np.append(prom, a.mean())
a = generarArray(10000)
print("10000 Corridas: ",a.mean())
prom = np.append(prom, a.mean())

print("\nPromedio general: ", prom.mean())


```

```python
import random
import numpy as np


def generar(n):
    e = np.exp(1)
    e = e**(-3)

    array = []
    contador = 0

    for i in range(n):
        u = random.random()
        M = u
        while M >= e:
            u = random.random()
            contador += 1
            M *= u
        array.append(contador)
        contador = 0
    return max(array)


def generarArray(n):
    data = np.array([])
    for i in range(n):
        data = np.append(data,generar(10))
    return data


def media(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum/len(arr)


a = generarArray(100)
print("100 Corridas: ",a.mean())
prom = np.append(prom, a.mean())
a = generarArray(1000)
print("1000 Corridas: ",a.mean())
prom = np.append(prom, a.mean())
a = generarArray(10000)
print("10000 Corridas: ",a.mean())
prom = np.append(prom, a.mean())

print("\nPromedio general: ", prom.mean())
```

# Ejercicio 04
Indique y describa las variables que serian necesarias tomar en cuenta para construir
un modelo matemático que pueda ser utilizado para simular el trabajo de una secretaria
dentro de una empresa (con el objetivo de establecer la eficiencia de la misma). 

```python
import numpy as np
import math
import random


def obtenerNumero(rand, n):
    acum = 0
    for i in range(1, n+1):
        if(rand >= acum and rand < (1/n)+acum):
            return i
        acum += 1/n


def getLado():
    return obtenerNumero(random.random(), 6)


dado1 = np.array([])
dado2 = np.array([])
dado3 = np.array([])
dado4 = np.array([])
dado5 = np.array([])
for i in range(50000):

    dado1 = np.append(dado1, getLado())
    dado2 = np.append(dado2, getLado())
    dado3 = np.append(dado3, getLado())
    dado4 = np.append(dado4, getLado())
    dado5 = np.append(dado5, getLado())

print("Dado 1: ", dado1)
print("Dado 2: ", dado2)
print("Dado 3: ", dado3)
print("Dado 4: ", dado4)
print("Dado 5: ", dado5)
```

# Ejercicio 05

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plot

X = np.array([])
Y = np.array([])

for i in range(20000):
    U1 = random.random()
    U2 = random.random()

    R_2 = -np.log(U1)
    O = 2 * math.pi * U2
    R = math.sqrt(R_2)

        
    X = np.append(X,R * math.cos(O)) 
    Y = np.append(Y,R * math.sin(O)) 

print("Numeros X: ",X)
print("Numeros Y:",Y)


plot.hist(x=X)
plot.hist(x=Y)
plot.title('Histograma de numeros')
plot.show()
```
